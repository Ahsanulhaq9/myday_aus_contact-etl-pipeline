```python
#v1

# Import required libraries
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
import pandas as pd
import time

# Function to get text from an element identified by its relative XPath, returns a default value if the element is not found

def get_element_text(driver, xpath, default=""):
    try:
        return driver.find_element(By.XPATH, xpath).text
    except:
        return default

# Function to scrape data from the inner page        
def scrape_inner_page(driver):
    print("Scraping inner page...")
    return {
        'People Section': get_element_text(driver, '//*[@id="content-inner"]/section/div[2]/div[1]/ng-include[1]/div[1]/h2'),
        'Full Name': get_element_text(driver, '//*[@id="content-inner"]/section/div[2]/div[1]/ng-include[1]/div[2]/div[1]/div/div[1]/div[2]/span'),
        'Physical Address': get_element_text(driver, '//*[@id="content-inner"]/section/div[2]/div[1]/ng-include[1]/div[2]/div[1]/div/div[4]/div[2]/span'),
        'Labels': get_element_text(driver, '//*[@id="vm.contact.Id"]/a/span/span'),
        'Mobile': get_element_text(driver, '//*[@id="content-inner"]/section/div[2]/div[1]/ng-include[1]/div[2]/div[2]/div/div[3]/div[2]/pm-send-ad-hoc-message/span/a/span'),
        'Work': get_element_text(driver, '//*[@id="content-inner"]/section/div[2]/div[1]/ng-include[1]/div[2]/div[2]/div/div[2]/div[2]/span'),
        'Email': get_element_text(driver, '//*[@id="content-inner"]/section/div[2]/div[1]/ng-include[1]/div[2]/div[2]/div/div[4]/div[2]/pm-send-ad-hoc-message/span/a/span'),
        'Communication': get_element_text(driver, '//*[@id="content-inner"]/section/div[2]/div[1]/ng-include[1]/div[2]/div[2]/div/div[5]/div[2]/span'),
        'Salutation': get_element_text(driver, '//*[@id="content-inner"]/section/div[2]/div[1]/ng-include[1]/div[2]/div[1]/div/div[2]/div[2]/span '),
        'Postal address': get_element_text(driver, '//*[@id="content-inner"]/section/div[2]/div[1]/ng-include[1]/div[2]/div[1]/div/div[3]/div[2]/span '),
        'Mobile 2': get_element_text(driver, '//*[@id="content-inner"]/section/div[2]/div[1]/ng-include[1]/div[2]/div[2]/div/div[3]/div[2]/pm-send-ad-hoc-message/span/span'),
        'Tenant': get_element_text(driver, '//*[@id="content-inner"]/section/div[2]/div[1]/ng-include[5]/div/h2/text()'),
        'Property': get_element_text(driver, '//*[@id="content-inner"]/section/div[2]/div[1]/ng-include[5]/div/div/div[2]/div[1]/div/div[1]/div[2]/a'),
        'Rent Arrears': get_element_text(driver, '//*[@id="content-inner"]/section/div[2]/div[1]/ng-include[5]/div/div/div[2]/div[1]/div/div[2]/div[1]/div[2]/span'),
        'Rent': get_element_text(driver, '//*[@id="content-inner"]/section/div[2]/div[1]/ng-include[5]/div/div/div[2]/div[1]/div/div[2]/div[2]/div[2]/span'),
        'Period': get_element_text(driver, '//*[@id="content-inner"]/section/div[2]/div[1]/ng-include[5]/div/div/div[2]/div[1]/div/div[2]/div[3]/div[2]/span'),
        'Moved In': get_element_text(driver, '//*[@id="content-inner"]/section/div[2]/div[1]/ng-include[5]/div/div/div[2]/div[1]/div/div[2]/div[4]/div[2]/span'),
        'Moved Out': get_element_text(driver, '//*[@id="content-inner"]/section/div[2]/div[1]/ng-include[5]/div/div/div[2]/div[1]/div/div[2]/div[5]/div[2]/span'),
        'Bond': get_element_text(driver, '//*[@id="content-inner"]/section/div[2]/div[1]/ng-include[5]/div/div/div[2]/div[1]/div/div[2]/div[6]/div[2]/a'),
        'Arrears Automation': get_element_text(driver, '//*[@id="content-inner"]/section/div[2]/div[1]/ng-include[5]/div/div/div[2]/div[1]/div/div[2]/div[8]/div[2]/span'),
        'Tenant Folio Code': get_element_text(driver, '//*[@id="content-inner"]/section/div[2]/div[1]/ng-include[5]/div/div/div[2]/div[2]/div/div[1]/div[2]/a'),
        'Invoices Due': get_element_text(driver, '//*[@id="content-inner"]/section/div[2]/div[1]/ng-include[5]/div/div/div[2]/div[2]/div/div[2]/div[1]/div[2]/span'),
        'Paid To': get_element_text(driver, '//*[@id="content-inner"]/section/div[2]/div[1]/ng-include[5]/div/div/div[2]/div[2]/div/div[2]/div[2]/div[2]/span'),
        'Eff. Paid To': get_element_text(driver, '//*[@id="content-inner"]/section/div[2]/div[1]/ng-include[5]/div/div/div[2]/div[2]/div/div[2]/div[4]/div[2]/span'),
        'Agreement': get_element_text(driver, '//*[@id="content-inner"]/section/div[2]/div[1]/ng-include[5]/div/div/div[2]/div[2]/div/div[2]/div[5]/div[2]/a'),
        'Periodic Tenancy': get_element_text(driver, '//*[@id="content-inner"]/section/div[2]/div[1]/ng-include[5]/div/div/div[2]/div[2]/div/div[2]/div[6]/div[2]/span'),
        'Deposits': get_element_text(driver, '//*[@id="content-inner"]/section/div[2]/div[1]/ng-include[5]/div/div/div[2]/div[2]/div/div[2]/div[7]/div[2]/span'),
        'Bank Reference': get_element_text(driver, '//*[@id="content-inner"]/section/div[2]/div[1]/ng-include[5]/div/div/div[2]/div[2]/div/div[2]/div[8]/div[2]/span'),
        'Receipt Warning': get_element_text(driver, '//*[@id="content-inner"]/section/div[2]/div[1]/ng-include[5]/div/div/div[2]/div[9]/div[2]/span'),
        'Supplier Part': get_element_text(driver, '//*[@id="content-inner"]/section/div[2]/div[1]/ng-include[6]/div/h2/text()'),
        'Company': get_element_text(driver, '//*[@id="content-inner"]/section/div[2]/div[1]/ng-include[6]/div/div/div[2]/div[1]/div/div[1]/div[2]/span'),
        'Bill Account': get_element_text(driver, '//*[@id="content-inner"]/section/div[2]/div[1]/ng-include[6]/div/div/div[2]/div[1]/div/div[2]/div[2]/span'),
        'Tenant Invoice Account': get_element_text(driver, '//*[@id="content-inner"]/section/div[2]/div[1]/ng-include[6]/div/div/div[2]/div[1]/div/div[3]'),
        'Website': get_element_text(driver, '//*[@id="content-inner"]/section/div[2]/div[1]/ng-include[6]/div/div/div[2]/div[1]/div/div[4]/div[2]/span'),
        'ABN': get_element_text(driver, '//*[@id="content-inner"]/section/div[2]/div[1]/ng-include[6]/div/div/div[2]/div[1]/div/div[5]/div[2]/span'),
        'Folio Code (Supplier)': get_element_text(driver, '//*[@id="content-inner"]/section/div[2]/div[1]/ng-include[6]/div/div/div[2]/div[2]/div/div[1]/div[2]/a'),
        'Balances': get_element_text(driver, '//*[@id="content-inner"]/section/div[2]/div[1]/ng-include[6]/div/div/div[2]/div[2]/div/div[2]/div[2]/span'),
        'Bills Pending': get_element_text(driver, '//*[@id="content-inner"]/section/div[2]/div[1]/ng-include[6]/div/div/div[2]/div[2]/div/div[3]/div[2]/span'),
        'Invoices Pending': get_element_text(driver, '//*[@id="content-inner"]/section/div[2]/div[1]/ng-include[6]/div/div/div[2]/div[2]/div/div[4]/div[2]/span'),
        'Bill Priority': get_element_text(driver, '//*[@id="content-inner"]/section/div[2]/div[1]/ng-include[6]/div/div/div[2]/div[2]/div/div[5]/div[2]/span'),
        'Payment Methods': get_element_text(driver, '//*[@id="content-inner"]/section/div[2]/div[1]/ng-include[6]/div/div/div[2]/div[2]/div/div[6]/div[2]/a'),
        'Seller': get_element_text(driver, '//*[@id="content-inner"]/section/div[2]/div[1]/ng-include[2]/div/h2/text()'),
        'Property (Buyer)': get_element_text(driver, '//*[@id="content-inner"]/section/div[2]/div[1]/ng-include[2]/div/div/div/div[2]/div[1]/div/div[1]/div[2]/a'),
        'Buyer': get_element_text(driver, '//*[@id="content-inner"]/section/div[2]/div[1]/ng-include[2]/div/div/div/div[2]/div[1]/div/div[2]/div[2]/a'),
        'Asking Price': get_element_text(driver, '//*[@id="content-inner"]/section/div[2]/div[1]/ng-include[2]/div/div/div/div[2]/div[1]/div/div[3]/div[2]/span'),
        'Purchase Price': get_element_text(driver, '//*[@id="content-inner"]/section/div[2]/div[1]/ng-include[2]/div/div/div/div[2]/div[1]/div/div[4]/div[2]/span'),
        'Commission': get_element_text(driver, '//*[@id="content-inner"]/section/div[2]/div[1]/ng-include[2]/div/div/div/div[2]/div[1]/div/div[5]/div[2]/span'),
        'Balance (Buyer)': get_element_text(driver, '//*[@id="content-inner"]/section/div[2]/div[1]/ng-include[2]/div/div/div/div[2]/div[1]/div/div[6]/div[2]/span'),
        'Bills Due': get_element_text(driver, '//*[@id="content-inner"]/section/div[2]/div[1]/ng-include[2]/div/div/div/div[2]/div[1]/div/div[7]/div[2]/span'),
        'Folio Code (Buyer)': get_element_text(driver, '//*[@id="content-inner"]/section/div[2]/div[1]/ng-include[2]/div/div/div/div[2]/div[2]/div/div[1]/div[2]/a'),
        'Agreement (Buyer)': get_element_text(driver, '//*[@id="content-inner"]/section/div[2]/div[1]/ng-include[2]/div/div/div/div[2]/div[2]/div/div[2]/div[2]/span'),
        'Payment Methods (Buyer)': get_element_text(driver, '//*[@id="content-inner"]/section/div[2]/div[1]/ng-include[2]/div/div/div/div[2]/div[2]/div/div[3]/div[2]/a'),
        'Deposit Due': get_element_text(driver, '//*[@id="content-inner"]/section/div[2]/div[1]/ng-include[2]/div/div/div/div[2]/div[2]/div/div[4]/div[1]/div[2]/span'),
        'Contract Exchange': get_element_text(driver, '//*[@id="content-inner"]/section/div[2]/div[1]/ng-include[2]/div/div/div/div[2]/div[2]/div/div[4]/div[2]/div[2]/span'),
        'Settlement': get_element_text(driver, '//*[@id="content-inner"]/section/div[2]/div[1]/ng-include[2]/div/div/div/div[2]/div[2]/div/div[4]/div[3]/div[2]/span'),
        'Owner': get_element_text(driver, '//*[@id="content-inner"]/section/div[2]/div[1]/ng-include[2]/div/h2/text()'),
        'Owner Property': get_element_text(driver, '//*[@id="content-inner"]/section/div[2]/div[1]/ng-include[2]/div[1]/div/div/div[2]/div[1]/div/div[1]/div[2]/a'),
        'Owner Fee schedule': get_element_text(driver, '//*[@id="content-inner"]/section/div[2]/div[1]/ng-include[4]/div[1]/div/div[2]/div[1]/div/div[2]/div[2]/a'),
        'Owner Disbursements': get_element_text(driver, '//*[@id="content-inner"]/section/div[2]/div[1]/ng-include[4]/div[1]/div/div[2]/div[1]/div/div[3]/div[2]/div/span'),
        'Owner Next disbursement': get_element_text(driver, '//*[@id="content-inner"]/section/div[2]/div[1]/ng-include[4]/div[1]/div/div[2]/div[1]/div/div[4]/div[2]/span'),
        'Owner Folio code': get_element_text(driver, '//*[@id="content-inner"]/section/div[2]/div[1]/ng-include[4]/div[1]/div/div[2]/div[2]/div/div[1]/div[2]/a'),
        'Owner Balance': get_element_text(driver, '//*[@id="content-inner"]/section/div[2]/div[1]/ng-include[4]/div[1]/div/div[2]/div[2]/div/div[2]/div[2]/span'),
        'Owner Withhold amount': get_element_text(driver, '//*[@id="content-inner"]/section/div[2]/div[1]/ng-include[4]/div[1]/div/div[2]/div[2]/div/div[3]/div[2]/span'),
        'Owner Payment methods': get_element_text(driver, '//*[@id="content-inner"]/section/div[2]/div[1]/ng-include[4]/div[1]/div/div[2]/div[2]/div/div[4]/div[2]/a'),
        'Owner Property 2': get_element_text(driver, '//*[@id="content-inner"]/section/div[2]/div[1]/ng-include[4]/div/div/div[2]/div[1]/div/div[1]/div[2]/a'),
        'Owner Fee schedule 2': get_element_text(driver, '//*[@id="content-inner"]/section/div[2]/div[1]/ng-include[4]/div/div/div[2]/div[1]/div/div[2]/div[2]/a'),
        'Owner Disbursements 2': get_element_text(driver, '//*[@id="content-inner"]/section/div[2]/div[1]/ng-include[4]/div/div/div[2]/div[1]/div/div[3]/div[2]/div/span'),
        'Owner Next disbursement 2': get_element_text(driver, '//*[@id="content-inner"]/section/div[2]/div[1]/ng-include[4]/div/div/div[2]/div[1]/div/div[4]/div[2]/span'),
        'Owner Folio code 2': get_element_text(driver, '//*[@id="content-inner"]/section/div[2]/div[1]/ng-include[4]/div/div/div[2]/div[2]/div/div[1]/div[2]/a'),
        'Owner Balance 2': get_element_text(driver, '//*[@id="content-inner"]/section/div[2]/div[1]/ng-include[4]/div/div/div[2]/div[2]/div/div[2]/div[2]/span'),
        'Owner Withhold amount 2': get_element_text(driver, '//*[@id="content-inner"]/section/div[2]/div[1]/ng-include[4]/div/div/div[2]/div[2]/div/div[3]/div[2]/span'),
        'Owner Payment methods 2': get_element_text(driver, '//*[@id="content-inner"]/section/div[2]/div[1]/ng-include[4]/div/div/div[2]/div[2]/div/div[4]/div[2]/a'),
    }

# Function to scrape data from the folio page
def scrape_folio_page(driver):
    print("Scraping folio page...")
    return {
        'Folio': get_element_text(driver, '//*[@id="content-inner"]/section/div[1]/h1'),
        'MePay Pending': get_element_text(driver, '//*[@id="content-inner"]/section/div[1]/div/div[2]/div/ng-include/div[1]/div[2]/span'),
        'Paid To': get_element_text(driver, '//*[@id="content-inner"]/section/div[1]/div/div[2]/div/ng-include/div[2]/div[2]/span'),
        'Part Paid': get_element_text(driver, '//*[@id="content-inner"]/section/div[1]/div/div[2]/div/ng-include/div[3]/div[2]/span'),
        'Eff. Paid To': get_element_text(driver, '//*[@id="content-inner"]/section/div[1]/div/div[2]/div/ng-include/div[4]/div[2]/span'),
        'Deposits': get_element_text(driver, '//*[@id="content-inner"]/section/div[1]/div/div[2]/div/ng-include/div[5]/div[2]/span'),
        'Uncleared': get_element_text(driver, '//*[@id="content-inner"]/section/div[1]/div/div[2]/div/ng-include/div[6]/div[2]/span'),
        'Bond Held': get_element_text(driver, '//*[@id="content-inner"]/section/div[1]/div/div[2]/div/ng-include/div[7]/div[2]/span'),
        'Owner Balance': get_element_text(driver, '//*[@id="content-inner"]/section/div[1]/div/div[2]/div/ng-include/div[8]/div[2]/span')
    }


# Function to scrape data from the transactions page
def scrape_transactions(driver):
    print("Scraping transactions...")
    transactions_data = []
    transaction_table = driver.find_element(By.XPATH, '//*[contains(@id, "DataTables_Table")]/tbody')
    rows = transaction_table.find_elements(By.TAG_NAME, 'tr')
    for row in rows:
        audit = get_element_text(row, './td[1]', "")
        date = get_element_text(row, './td[2]', "")
        ref = get_element_text(row, './td[3]/span', "")
        type_ = get_element_text(row, './td[4]', "")
        property_ = get_element_text(row, './td[5]/span', "")
        summary = get_element_text(row, './td[6]/span', "")
        debit = get_element_text(row, './td[7]', "")
        credit = get_element_text(row, './td[8]', "")
        transactions_data.append([audit, date, ref, type_, property_, summary, debit, credit])
    return transactions_data

# Function to transform transactions DataFrame to a single row DataFrame with unique column names
def transform_to_single_row(df):
    columns = df.columns
    data = {}
    for i, row in df.iterrows():
        for col in columns:
            data[f'{col}{i+1}'] = row[col]
    return pd.DataFrame([data])


# Function to click the folio code link
def click_folio_code(driver):
    folio_code_link = None
    try:
        # Try Tenant folio code first
        folio_code_link = driver.find_element(By.XPATH, '//*[@id="content-inner"]/section/div[2]/div[1]/ng-include[5]/div/div/div[2]/div[2]/div/div[1]/div[2]/a')
    except:
        try:
            # Try Supplier folio code
            folio_code_link = driver.find_element(By.XPATH, '//*[@id="content-inner"]/section/div[2]/div[1]/ng-include[6]/div/div/div[2]/div[2]/div/div[1]/div[2]/a')
        except:
            try:
                # Try Seller folio code
                folio_code_link = driver.find_element(By.XPATH, '//*[@id="content-inner"]/section/div[2]/div[1]/ng-include[2]/div/div/div/div[2]/div[2]/div/div[1]/div[2]/a')
            except:
                try:
                    # Try Owner folio code
                    folio_code_link = driver.find_element(By.XPATH, '//*[@id="content-inner"]/section/div[2]/div[1]/ng-include[4]/div[1]/div/div[2]/div[2]/div/div[1]/div[2]/a')
                except:
                    folio_code_link = None

    if folio_code_link:
        driver.execute_script("arguments[0].click();", folio_code_link)
        return True
    else:
        return False


# Function to scrape data from the contacts page
def scrape_contacts_page(driver):
    print("Extracting rows from the contacts table...")
    # Get the rows from the contacts table
    contacts_table = driver.find_element(By.XPATH, '//*[contains(@id, "DataTables_Table")]/tbody')
    rows = contacts_table.find_elements(By.TAG_NAME, 'tr')

    data = []

    # Extract data from each row
    for row in rows:
        reference = get_element_text(row, './td[2]/span/a')
        primary_person = get_element_text(row, './td[3]/span')
        phone = get_element_text(row, './td[4]/span')
        roles = get_element_text(row, './td[5]/span')
        labels = get_element_text(row, './td[6]')
        data.append([reference, primary_person, phone, roles, labels])

    # Create a DataFrame
    contacts_df = pd.DataFrame(data, columns=['Reference', 'Primary Person', 'Phone', 'Roles', 'Labels'])
    return contacts_df, rows

# Initialize Chrome WebDriver
print("Initializing Chrome WebDriver...")
driver = webdriver.Chrome()
wait = WebDriverWait(driver, 10)

# Go to the URL
print("Navigating to the URL...")
driver.get('https://manager.propertyme.com/#/property/list')

# Wait for the page to load completely
time.sleep(3)

print("Inputting email address...")
# Input the email address
email_input = wait.until(EC.presence_of_element_located((By.XPATH, '//*[@id="pme-input-0"]')))
email_input.send_keys('mannan1985on@gmail.com')

print("Inputting password...")
# Input the password
password_input = wait.until(EC.presence_of_element_located((By.XPATH, '//*[@id="pme-input-1"]')))
password_input.send_keys('Property1234')

print("Clicking login button...")
# Use JavaScript to click on the login button
login_button = wait.until(EC.element_to_be_clickable((By.XPATH, '/html/body/pme-root/pme-feature-shell/main/section/pme-feature-login/section/form/button/span')))
driver.execute_script("arguments[0].click();", login_button)

# Wait for the two-factor authentication page to load
time.sleep(5)  # Adjust this sleep time as needed

print("Please enter the authentication code.")
# Take input from user for authentication code
auth_code = input("Please enter the authentication code: ")

print("Inputting authentication code...")
# Input the authentication code
auth_code_input = wait.until(EC.presence_of_element_located((By.XPATH, '//*[@id="pme-input-2"]')))
auth_code_input.send_keys(auth_code)

print("Clicking on the checkbox to trust this device for 7 days...")
# Click on the checkbox to trust this device for 7 days
checkbox = wait.until(EC.element_to_be_clickable((By.XPATH, '//*[@id="pme-checkbox-0"]/label/div[1]')))
driver.execute_script("arguments[0].click();", checkbox)

print("Clicking verify button...")
# Click on the verify button
verify_button = wait.until(EC.element_to_be_clickable((By.XPATH, '/html/body/pme-root/pme-feature-shell/main/section/pme-feature-login-mfa/section/form/button')))
driver.execute_script("arguments[0].click();", verify_button)

# Wait for the next page to load after verification
time.sleep(15)  # Adjust this sleep time as needed

print("Clicking on Contacts...")
# Click on Contacts
contacts_button = wait.until(EC.element_to_be_clickable((By.XPATH, '//*[@id="sidebar"]/div[1]/div/div[1]/ul/li[5]/a')))
#contacts_button = wait.until(EC.element_to_be_clickable((By.XPATH, '//*[@id="sidebar"]/div[1]/div/div[1]/ul/li[5]')))
driver.execute_script("arguments[0].click();", contacts_button)

# Wait for the Contacts page to load
time.sleep(5)  # Adjust this sleep time as needed

# Lists to hold data from different pages
all_combined_dfs = []
inner_pages = []
folio_pages = []
transaction_pages = []

# Function to process each row and scrape required data
def process_row(driver, row_index, contacts_df, wait):
    print(f"Processing row {row_index}...")
    reference_link = wait.until(EC.element_to_be_clickable((By.XPATH, f'//*[contains(@id, "DataTables_Table")]/tbody/tr[{row_index}]/td[2]/span/a')))
    # Open reference link in a new tab
    driver.execute_script("window.open(arguments[0].href, '_blank');", reference_link)
    driver.switch_to.window(driver.window_handles[1])
    time.sleep(5)

    inner_data = scrape_inner_page(driver)
    inner_df = pd.DataFrame([inner_data])
    inner_df.columns = ['Inner_' + col for col in inner_df.columns]
    inner_pages.append(inner_df)
    
    folio_code_found = click_folio_code(driver)
    if folio_code_found:
        time.sleep(5)
        folio_data = scrape_folio_page(driver)
        folio_df = pd.DataFrame([folio_data])
    else:
        folio_df = pd.DataFrame([{}])
    folio_df.columns = ['Folio_' + col for col in folio_df.columns]
    folio_pages.append(folio_df)

    if folio_code_found:
        all_transactions_tab = wait.until(EC.element_to_be_clickable((By.XPATH, '//*[@id="grid-tab-nav-tabs"]/li[2]/a')))
        driver.execute_script("arguments[0].click();", all_transactions_tab)
        time.sleep(5)
        transactions_data = scrape_transactions(driver)
        transactions_df = pd.DataFrame(transactions_data, columns=['Audit', 'Date', 'Reference', 'Type', 'Property', 'Summary', 'Debit', 'Credit'])
        transformed_df = transform_to_single_row(transactions_df)
    else:
        transactions_data = [{'Audit': '', 'Date': '', 'Reference': '', 'Type': '', 'Property': '', 'Summary': '', 'Debit': '', 'Credit': ''}]
        transformed_df = pd.DataFrame(transactions_data)
    transformed_df.columns = ['Trans_' + col for col in transformed_df.columns]
    transaction_pages.append(transformed_df)
    
    combined_df = pd.concat([contacts_df.iloc[[row_index - 1]].reset_index(drop=True), inner_df.reset_index(drop=True), folio_df.reset_index(drop=True), transformed_df.reset_index(drop=True)], axis=1)

    # Close the tab and switch back to the first tab
    driver.close()
    driver.switch_to.window(driver.window_handles[0])
    return combined_df

# Initialize page number
page_number = 1

# Try block to ensure resources are released properly
try:
    while True:
        print(f"Scraping page {page_number}...")
        contacts_df, rows = scrape_contacts_page(driver)
        contacts_df.to_excel(f'contacts_data_page_{page_number}.xlsx', index=False)

        # Process each row in the current page
        for row_index in range(1, len(rows) + 1):
            try:
                combined_df = process_row(driver, row_index, contacts_df, wait)
                all_combined_dfs.append(combined_df)
                print(f"Returning to Contacts page after processing row {row_index} on page {page_number}...")
            except Exception as e:
                print(f"Error processing row {row_index} on page {page_number}: {e}")
                continue

        # Click on the next page button if available
        try:
            print("Clicking on the next page button...")
            next_button = wait.until(EC.element_to_be_clickable((By.XPATH, '//*[contains(@id, "DataTables_Table") and contains(@id, "_paginate")]/ul/li[@class="next"]/a')))
            driver.execute_script("arguments[0].click();", next_button)
            print("Waiting for the next page to load...")
            page_number += 1
            time.sleep(5)  # Adjust this sleep time as needed
        except Exception as e:
            print("No more pages or an error occurred:", e)
            break
finally:
    # Save all collected data to Excel files
    print("Saving inner pages DataFrame to an Excel file...")
    inner_pages_df = pd.concat(inner_pages, axis=0, ignore_index=True)
    inner_pages_df.to_excel('inner_pages_data.xlsx', index=False)

    print("Saving folio pages DataFrame to an Excel file...")
    folio_pages_df = pd.concat(folio_pages, axis=0, ignore_index=True)
    folio_pages_df.to_excel('folio_pages_data.xlsx', index=False)

    print("Saving transactions DataFrame to an Excel file...")
    transaction_pages_df = pd.concat(transaction_pages, axis=0, ignore_index=True)
    transaction_pages_df.to_excel('transactions_data.xlsx', index=False)

    print("Concatenating all combined DataFrames...")
    final_combined_df = pd.concat(all_combined_dfs, axis=0, ignore_index=True)

    print("Saving final combined DataFrame to a new Excel file...")
    final_combined_df.to_excel('final_combined_data.xlsx', index=False)

    print("Closing the driver...")
    driver.quit()
```

```python
#v2 ,resumes after any error if above code gives and sxtracts data after that

from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
import pandas as pd
import time
import os

def get_element_text(driver, xpath, default=""):
    try:
        text = driver.find_element(By.XPATH, xpath).text
        return ''.join(c for c in text if c.isprintable())
    except:
        return default

def scrape_inner_page(driver):
    print("Scraping inner page...")
    return {
        'People Section': get_element_text(driver, '//*[@id="content-inner"]/section/div[2]/div[1]/ng-include[1]/div[1]/h2'),
        'Full Name': get_element_text(driver, '//*[@id="content-inner"]/section/div[2]/div[1]/ng-include[1]/div[2]/div[1]/div/div[1]/div[2]/span'),
        'Physical Address': get_element_text(driver, '//*[@id="content-inner"]/section/div[2]/div[1]/ng-include[1]/div[2]/div[1]/div/div[4]/div[2]/span'),
        'Labels': get_element_text(driver, '//*[@id="vm.contact.Id"]/a/span/span'),
        'Mobile': get_element_text(driver, '//*[@id="content-inner"]/section/div[2]/div[1]/ng-include[1]/div[2]/div[2]/div/div[3]/div[2]/pm-send-ad-hoc-message/span/a/span'),
        'Work': get_element_text(driver, '//*[@id="content-inner"]/section/div[2]/div[1]/ng-include[1]/div[2]/div[2]/div/div[2]/div[2]/span'),
        'Email': get_element_text(driver, '//*[@id="content-inner"]/section/div[2]/div[1]/ng-include[1]/div[2]/div[2]/div/div[4]/div[2]/pm-send-ad-hoc-message/span/a/span'),
        'Communication': get_element_text(driver, '//*[@id="content-inner"]/section/div[2]/div[1]/ng-include[1]/div[2]/div[2]/div/div[5]/div[2]/span'),
        'Salutation': get_element_text(driver, '//*[@id="content-inner"]/section/div[2]/div[1]/ng-include[1]/div[2]/div[1]/div/div[2]/div[2]/span '),
        'Postal address': get_element_text(driver, '//*[@id="content-inner"]/section/div[2]/div[1]/ng-include[1]/div[2]/div[1]/div/div[3]/div[2]/span '),
        'Mobile 2': get_element_text(driver, '//*[@id="content-inner"]/section/div[2]/div[1]/ng-include[1]/div[2]/div[2]/div/div[3]/div[2]/pm-send-ad-hoc-message/span/span'),
        'Tenant': get_element_text(driver, '//*[@id="content-inner"]/section/div[2]/div[1]/ng-include[5]/div/h2/text()'),
        'Property': get_element_text(driver, '//*[@id="content-inner"]/section/div[2]/div[1]/ng-include[5]/div/div/div[2]/div[1]/div/div[1]/div[2]/a'),
        'Rent Arrears': get_element_text(driver, '//*[@id="content-inner"]/section/div[2]/div[1]/ng-include[5]/div/div/div[2]/div[1]/div/div[2]/div[1]/div[2]/span'),
        'Rent': get_element_text(driver, '//*[@id="content-inner"]/section/div[2]/div[1]/ng-include[5]/div/div/div[2]/div[1]/div/div[2]/div[2]/div[2]/span'),
        'Period': get_element_text(driver, '//*[@id="content-inner"]/section/div[2]/div[1]/ng-include[5]/div/div/div[2]/div[1]/div/div[2]/div[3]/div[2]/span'),
        'Moved In': get_element_text(driver, '//*[@id="content-inner"]/section/div[2]/div[1]/ng-include[5]/div/div/div[2]/div[1]/div/div[2]/div[4]/div[2]/span'),
        'Moved Out': get_element_text(driver, '//*[@id="content-inner"]/section/div[2]/div[1]/ng-include[5]/div/div/div[2]/div[1]/div/div[2]/div[5]/div[2]/span'),
        'Bond': get_element_text(driver, '//*[@id="content-inner"]/section/div[2]/div[1]/ng-include[5]/div/div/div[2]/div[1]/div/div[2]/div[6]/div[2]/a'),
        'Arrears Automation': get_element_text(driver, '//*[@id="content-inner"]/section/div[2]/div[1]/ng-include[5]/div/div/div[2]/div[1]/div/div[2]/div[8]/div[2]/span'),
        'Tenant Folio Code': get_element_text(driver, '//*[@id="content-inner"]/section/div[2]/div[1]/ng-include[5]/div/div/div[2]/div[2]/div/div[1]/div[2]/a'),
        'Invoices Due': get_element_text(driver, '//*[@id="content-inner"]/section/div[2]/div[1]/ng-include[5]/div/div/div[2]/div[2]/div/div[2]/div[1]/div[2]/span'),
        'Paid To': get_element_text(driver, '//*[@id="content-inner"]/section/div[2]/div[1]/ng-include[5]/div/div/div[2]/div[2]/div/div[2]/div[2]/div[2]/span'),
        'Eff. Paid To': get_element_text(driver, '//*[@id="content-inner"]/section/div[2]/div[1]/ng-include[5]/div/div/div[2]/div[2]/div/div[2]/div[4]/div[2]/span'),
        'Agreement': get_element_text(driver, '//*[@id="content-inner"]/section/div[2]/div[1]/ng-include[5]/div/div/div[2]/div[2]/div/div[2]/div[5]/div[2]/a'),
        'Periodic Tenancy': get_element_text(driver, '//*[@id="content-inner"]/section/div[2]/div[1]/ng-include[5]/div/div/div[2]/div[2]/div/div[2]/div[6]/div[2]/span'),
        'Deposits': get_element_text(driver, '//*[@id="content-inner"]/section/div[2]/div[1]/ng-include[5]/div/div/div[2]/div[2]/div/div[2]/div[7]/div[2]/span'),
        'Bank Reference': get_element_text(driver, '//*[@id="content-inner"]/section/div[2]/div[1]/ng-include[5]/div/div/div[2]/div[2]/div/div[2]/div[8]/div[2]/span'),
        'Receipt Warning': get_element_text(driver, '//*[@id="content-inner"]/section/div[2]/div[1]/ng-include[5]/div/div/div[2]/div[9]/div[2]/span'),
        'Supplier Part': get_element_text(driver, '//*[@id="content-inner"]/section/div[2]/div[1]/ng-include[6]/div/h2/text()'),
        'Company': get_element_text(driver, '//*[@id="content-inner"]/section/div[2]/div[1]/ng-include[6]/div/div/div[2]/div[1]/div/div[1]/div[2]/span'),
        'Bill Account': get_element_text(driver, '//*[@id="content-inner"]/section/div[2]/div[1]/ng-include[6]/div/div/div[2]/div[1]/div/div[2]/div[2]/span'),
        'Tenant Invoice Account': get_element_text(driver, '//*[@id="content-inner"]/section/div[2]/div[1]/ng-include[6]/div/div/div[2]/div[1]/div/div[3]'),
        'Website': get_element_text(driver, '//*[@id="content-inner"]/section/div[2]/div[1]/ng-include[6]/div/div/div[2]/div[1]/div/div[4]/div[2]/span'),
        'ABN': get_element_text(driver, '//*[@id="content-inner"]/section/div[2]/div[1]/ng-include[6]/div/div/div[2]/div[1]/div/div[5]/div[2]/span'),
        'Folio Code (Supplier)': get_element_text(driver, '//*[@id="content-inner"]/section/div[2]/div[1]/ng-include[6]/div/div/div[2]/div[2]/div/div[1]/div[2]/a'),
        'Balances': get_element_text(driver, '//*[@id="content-inner"]/section/div[2]/div[1]/ng-include[6]/div/div/div[2]/div[2]/div/div[2]/div[2]/span'),
        'Bills Pending': get_element_text(driver, '//*[@id="content-inner"]/section/div[2]/div[1]/ng-include[6]/div/div/div[2]/div[2]/div/div[3]/div[2]/span'),
        'Invoices Pending': get_element_text(driver, '//*[@id="content-inner"]/section/div[2]/div[1]/ng-include[6]/div/div/div[2]/div[2]/div/div[4]/div[2]/span'),
        'Bill Priority': get_element_text(driver, '//*[@id="content-inner"]/section/div[2]/div[1]/ng-include[6]/div/div/div[2]/div[2]/div/div[5]/div[2]/span'),
        'Payment Methods': get_element_text(driver, '//*[@id="content-inner"]/section/div[2]/div[1]/ng-include[6]/div/div/div[2]/div[2]/div/div[6]/div[2]/a'),
        'Seller': get_element_text(driver, '//*[@id="content-inner"]/section/div[2]/div[1]/ng-include[2]/div/h2/text()'),
        'Property (Buyer)': get_element_text(driver, '//*[@id="content-inner"]/section/div[2]/div[1]/ng-include[2]/div/div/div/div[2]/div[1]/div/div[1]/div[2]/a'),
        'Buyer': get_element_text(driver, '//*[@id="content-inner"]/section/div[2]/div[1]/ng-include[2]/div/div/div/div[2]/div[1]/div/div[2]/div[2]/a'),
        'Asking Price': get_element_text(driver, '//*[@id="content-inner"]/section/div[2]/div[1]/ng-include[2]/div/div/div/div[2]/div[1]/div/div[3]/div[2]/span'),
        'Purchase Price': get_element_text(driver, '//*[@id="content-inner"]/section/div[2]/div[1]/ng-include[2]/div/div/div/div[2]/div[1]/div/div[4]/div[2]/span'),
        'Commission': get_element_text(driver, '//*[@id="content-inner"]/section/div[2]/div[1]/ng-include[2]/div/div/div/div[2]/div[1]/div/div[5]/div[2]/span'),
        'Balance (Buyer)': get_element_text(driver, '//*[@id="content-inner"]/section/div[2]/div[1]/ng-include[2]/div/div/div/div[2]/div[1]/div/div[6]/div[2]/span'),
        'Bills Due': get_element_text(driver, '//*[@id="content-inner"]/section/div[2]/div[1]/ng-include[2]/div/div/div/div[2]/div[1]/div/div[7]/div[2]/span'),
        'Folio Code (Buyer)': get_element_text(driver, '//*[@id="content-inner"]/section/div[2]/div[1]/ng-include[2]/div/div/div/div[2]/div[2]/div/div[1]/div[2]/a'),
        'Agreement (Buyer)': get_element_text(driver, '//*[@id="content-inner"]/section/div[2]/div[1]/ng-include[2]/div/div/div/div[2]/div[2]/div/div[2]/div[2]/span'),
        'Payment Methods (Buyer)': get_element_text(driver, '//*[@id="content-inner"]/section/div[2]/div[1]/ng-include[2]/div/div/div/div[2]/div[2]/div/div[3]/div[2]/a'),
        'Deposit Due': get_element_text(driver, '//*[@id="content-inner"]/section/div[2]/div[1]/ng-include[2]/div/div/div/div[2]/div[2]/div/div[4]/div[1]/div[2]/span'),
        'Contract Exchange': get_element_text(driver, '//*[@id="content-inner"]/section/div[2]/div[1]/ng-include[2]/div/div/div/div[2]/div[2]/div/div[4]/div[2]/div[2]/span'),
        'Settlement': get_element_text(driver, '//*[@id="content-inner"]/section/div[2]/div[1]/ng-include[2]/div/div/div/div[2]/div[2]/div/div[4]/div[3]/div[2]/span'),
        'Owner': get_element_text(driver, '//*[@id="content-inner"]/section/div[2]/div[1]/ng-include[2]/div/h2/text()'),
        'Owner Property': get_element_text(driver, '//*[@id="content-inner"]/section/div[2]/div[1]/ng-include[2]/div[1]/div/div/div[2]/div[1]/div/div[1]/div[2]/a'),
        'Owner Fee schedule': get_element_text(driver, '//*[@id="content-inner"]/section/div[2]/div[1]/ng-include[4]/div[1]/div/div[2]/div[1]/div/div[2]/div[2]/a'),
        'Owner Disbursements': get_element_text(driver, '//*[@id="content-inner"]/section/div[2]/div[1]/ng-include[4]/div[1]/div/div[2]/div[1]/div/div[3]/div[2]/div/span'),
        'Owner Next disbursement': get_element_text(driver, '//*[@id="content-inner"]/section/div[2]/div[1]/ng-include[4]/div[1]/div/div[2]/div[1]/div/div[4]/div[2]/span'),
        'Owner Folio code': get_element_text(driver, '//*[@id="content-inner"]/section/div[2]/div[1]/ng-include[4]/div[1]/div/div[2]/div[2]/div/div[1]/div[2]/a'),
        'Owner Balance': get_element_text(driver, '//*[@id="content-inner"]/section/div[2]/div[1]/ng-include[4]/div[1]/div/div[2]/div[2]/div/div[2]/div[2]/span'),
        'Owner Withhold amount': get_element_text(driver, '//*[@id="content-inner"]/section/div[2]/div[1]/ng-include[4]/div[1]/div/div[2]/div[2]/div/div[3]/div[2]/span'),
        'Owner Payment methods': get_element_text(driver, '//*[@id="content-inner"]/section/div[2]/div[1]/ng-include[4]/div[1]/div/div[2]/div[2]/div/div[4]/div[2]/a'),
        'Owner Property 2': get_element_text(driver, '//*[@id="content-inner"]/section/div[2]/div[1]/ng-include[4]/div/div/div[2]/div[1]/div/div[1]/div[2]/a'),
        'Owner Fee schedule 2': get_element_text(driver, '//*[@id="content-inner"]/section/div[2]/div[1]/ng-include[4]/div/div/div[2]/div[1]/div/div[2]/div[2]/a'),
        'Owner Disbursements 2': get_element_text(driver, '//*[@id="content-inner"]/section/div[2]/div[1]/ng-include[4]/div/div/div[2]/div[1]/div/div[3]/div[2]/div/span'),
        'Owner Next disbursement 2': get_element_text(driver, '//*[@id="content-inner"]/section/div[2]/div[1]/ng-include[4]/div/div/div[2]/div[1]/div/div[4]/div[2]/span'),
        'Owner Folio code 2': get_element_text(driver, '//*[@id="content-inner"]/section/div[2]/div[1]/ng-include[4]/div/div/div[2]/div[2]/div/div[1]/div[2]/a'),
        'Owner Balance 2': get_element_text(driver, '//*[@id="content-inner"]/section/div[2]/div[1]/ng-include[4]/div/div/div[2]/div[2]/div/div[2]/div[2]/span'),
        'Owner Withhold amount 2': get_element_text(driver, '//*[@id="content-inner"]/section/div[2]/div[1]/ng-include[4]/div/div/div[2]/div[2]/div/div[3]/div[2]/span'),
        'Owner Payment methods 2': get_element_text(driver, '//*[@id="content-inner"]/section/div[2]/div[1]/ng-include[4]/div/div/div[2]/div[2]/div/div[4]/div[2]/a'),
    }

def scrape_folio_page(driver):
    print("Scraping folio page...")
    return {
        'Folio': get_element_text(driver, '//*[@id="content-inner"]/section/div[1]/h1'),
        'MePay Pending': get_element_text(driver, '//*[@id="content-inner"]/section/div[1]/div/div[2]/div/ng-include/div[1]/div[2]/span'),
        'Paid To': get_element_text(driver, '//*[@id="content-inner"]/section/div[1]/div/div[2]/div/ng-include/div[2]/div[2]/span'),
        'Part Paid': get_element_text(driver, '//*[@id="content-inner"]/section/div[1]/div/div[2]/div/ng-include/div[3]/div[2]/span'),
        'Eff. Paid To': get_element_text(driver, '//*[@id="content-inner"]/section/div[1]/div/div[2]/div/ng-include/div[4]/div[2]/span'),
        'Deposits': get_element_text(driver, '//*[@id="content-inner"]/section/div[1]/div/div[2]/div/ng-include/div[5]/div[2]/span'),
        'Uncleared': get_element_text(driver, '//*[@id="content-inner"]/section/div[1]/div/div[2]/div/ng-include/div[6]/div[2]/span'),
        'Bond Held': get_element_text(driver, '//*[@id="content-inner"]/section/div[1]/div/div[2]/div/ng-include/div[7]/div[2]/span'),
        'Owner Balance': get_element_text(driver, '//*[@id="content-inner"]/section/div[1]/div/div[2]/div/ng-include/div[8]/div[2]/span')
    }

def scrape_transactions(driver):
    print("Scraping transactions...")
    transactions_data = []
    transaction_table = driver.find_element(By.XPATH, '//*[contains(@id, "DataTables_Table")]/tbody')
    rows = transaction_table.find_elements(By.TAG_NAME, 'tr')
    for row in rows:
        audit = get_element_text(row, './td[1]', "")
        date = get_element_text(row, './td[2]', "")
        ref = get_element_text(row, './td[3]/span', "")
        type_ = get_element_text(row, './td[4]', "")
        property_ = get_element_text(row, './td[5]/span', "")
        summary = get_element_text(row, './td[6]/span', "")
        debit = get_element_text(row, './td[7]', "")
        credit = get_element_text(row, './td[8]', "")
        transactions_data.append([audit, date, ref, type_, property_, summary, debit, credit])
    return transactions_data

def transform_to_single_row(df):
    columns = df.columns
    data = {}
    for i, row in df.iterrows():
        for col in columns:
            data[f'{col}{i+1}'] = row[col]
    return pd.DataFrame([data])

def click_folio_code(driver):
    folio_code_link = None
    try:
        # Try Tenant folio code first
        folio_code_link = driver.find_element(By.XPATH, '//*[@id="content-inner"]/section/div[2]/div[1]/ng-include[5]/div/div/div[2]/div[2]/div/div[1]/div[2]/a')
    except:
        try:
            # Try Supplier folio code
            folio_code_link = driver.find_element(By.XPATH, '//*[@id="content-inner"]/section/div[2]/div[1]/ng-include[6]/div/div/div[2]/div[2]/div/div[1]/div[2]/a')
        except:
            try:
                # Try Seller folio code
                folio_code_link = driver.find_element(By.XPATH, '//*[@id="content-inner"]/section/div[2]/div[1]/ng-include[2]/div/div/div/div[2]/div[2]/div/div[1]/div[2]/a')
            except:
                try:
                    # Try Owner folio code
                    folio_code_link = driver.find_element(By.XPATH, '//*[@id="content-inner"]/section/div[2]/div[1]/ng-include[4]/div[1]/div/div[2]/div[2]/div/div[1]/div[2]/a')
                except:
                    folio_code_link = None

    if folio_code_link:
        driver.execute_script("arguments[0].click();", folio_code_link)
        return True
    else:
        return False

def scrape_contacts_page(driver):
    print("Extracting rows from the contacts table...")
    # Get the rows from the contacts table
    contacts_table = driver.find_element(By.XPATH, '//*[contains(@id, "DataTables_Table")]/tbody')
    rows = contacts_table.find_elements(By.TAG_NAME, 'tr')

    data = []

    # Extract data from each row
    for row in rows:
        reference = get_element_text(row, './td[2]/span/a')
        primary_person = get_element_text(row, './td[3]/span')
        phone = get_element_text(row, './td[4]/span')
        roles = get_element_text(row, './td[5]/span')
        labels = get_element_text(row, './td[6]')
        data.append([reference, primary_person, phone, roles, labels])

    # Create a DataFrame
    contacts_df = pd.DataFrame(data, columns=['Reference', 'Primary Person', 'Phone', 'Roles', 'Labels'])
    return contacts_df, rows

def load_existing_data():
    if os.path.exists('final_combined_data.xlsx'):
        existing_data = pd.read_excel('final_combined_data.xlsx')
        return existing_data
    else:
        return pd.DataFrame()

# Initialize Chrome WebDriver
print("Initializing Chrome WebDriver...")
driver = webdriver.Chrome()
wait = WebDriverWait(driver, 10)

# Go to the URL
print("Navigating to the URL...")
driver.get('https://manager.propertyme.com/#/property/list')

# Wait for the page to load completely
time.sleep(3)

print("Inputting email address...")
# Input the email address
email_input = wait.until(EC.presence_of_element_located((By.XPATH, '//*[@id="pme-input-0"]')))
email_input.send_keys('mannan1985on@gmail.com')

print("Inputting password...")
# Input the password
password_input = wait.until(EC.presence_of_element_located((By.XPATH, '//*[@id="pme-input-1"]')))
password_input.send_keys('Property1234')

print("Clicking login button...")
# Use JavaScript to click on the login button
login_button = wait.until(EC.element_to_be_clickable((By.XPATH, '/html/body/pme-root/pme-feature-shell/main/section/pme-feature-login/section/form/button/span')))
driver.execute_script("arguments[0].click();", login_button)

# Wait for the two-factor authentication page to load
time.sleep(5)  # Adjust this sleep time as needed

print("Please enter the authentication code.")
# Take input from user for authentication code
auth_code = input("Please enter the authentication code: ")

print("Inputting authentication code...")
# Input the authentication code
auth_code_input = wait.until(EC.presence_of_element_located((By.XPATH, '//*[@id="pme-input-2"]')))
auth_code_input.send_keys(auth_code)

print("Clicking on the checkbox to trust this device for 7 days...")
# Click on the checkbox to trust this device for 7 days
checkbox = wait.until(EC.element_to_be_clickable((By.XPATH, '//*[@id="pme-checkbox-0"]/label/div[1]')))
driver.execute_script("arguments[0].click();", checkbox)

print("Clicking verify button...")
# Click on the verify button
verify_button = wait.until(EC.element_to_be_clickable((By.XPATH, '/html/body/pme-root/pme-feature-shell/main/section/pme-feature-login-mfa/section/form/button')))
driver.execute_script("arguments[0].click();", verify_button)

# Wait for the next page to load after verification
time.sleep(15)  # Adjust this sleep time as needed

print("Clicking on Contacts...")
# Click on Contacts
contacts_button = wait.until(EC.element_to_be_clickable((By.XPATH, '//*[@id="sidebar"]/div[1]/div/div[1]/ul/li[5]/a')))
driver.execute_script("arguments[0].click();", contacts_button)

# Wait for the Contacts page to load
time.sleep(5)  # Adjust this sleep time as needed

all_combined_dfs = []
inner_pages = []
folio_pages = []
transaction_pages = []

# Load existing data
existing_data = load_existing_data()

# Function to process each row and scrape required data
def process_row(driver, row_index, contacts_df, wait):
    print(f"Processing row {row_index}...")
    reference_link = wait.until(EC.element_to_be_clickable((By.XPATH, f'//*[contains(@id, "DataTables_Table")]/tbody/tr[{row_index}]/td[2]/span/a')))
    # Open reference link in a new tab
    driver.execute_script("window.open(arguments[0].href, '_blank');", reference_link)
    driver.switch_to.window(driver.window_handles[1])
    time.sleep(5)

    inner_data = scrape_inner_page(driver)
    inner_df = pd.DataFrame([inner_data])
    inner_df.columns = ['Inner_' + col for col in inner_df.columns]
    inner_pages.append(inner_df)
    
    folio_code_found = click_folio_code(driver)
    if folio_code_found:
        time.sleep(5)
        folio_data = scrape_folio_page(driver)
        folio_df = pd.DataFrame([folio_data])
    else:
        folio_df = pd.DataFrame([{}])
    folio_df.columns = ['Folio_' + col for col in folio_df.columns]
    folio_pages.append(folio_df)

    if folio_code_found:
        all_transactions_tab = wait.until(EC.element_to_be_clickable((By.XPATH, '//*[@id="grid-tab-nav-tabs"]/li[2]/a')))
        driver.execute_script("arguments[0].click();", all_transactions_tab)
        time.sleep(5)
        transactions_data = scrape_transactions(driver)
        transactions_df = pd.DataFrame(transactions_data, columns=['Audit', 'Date', 'Reference', 'Type', 'Property', 'Summary', 'Debit', 'Credit'])
        transformed_df = transform_to_single_row(transactions_df)
    else:
        transactions_data = [{'Audit': '', 'Date': '', 'Reference': '', 'Type': '', 'Property': '', 'Summary': '', 'Debit': '', 'Credit': ''}]
        transformed_df = pd.DataFrame(transactions_data)
    transformed_df.columns = ['Trans_' + col for col in transformed_df.columns]
    transaction_pages.append(transformed_df)
    
    combined_df = pd.concat([contacts_df.iloc[[row_index - 1]].reset_index(drop=True), inner_df.reset_index(drop=True), folio_df.reset_index(drop=True), transformed_df.reset_index(drop=True)], axis=1)

    # Close the tab and switch back to the first tab
    driver.close()
    driver.switch_to.window(driver.window_handles[0])
    return combined_df

page_number = 1

try:
    while True:
        print(f"Scraping page {page_number}...")
        contacts_df, rows = scrape_contacts_page(driver)
        contacts_df.to_excel(f'contacts_data_page_{page_number}.xlsx', index=False)

        for row_index in range(1, len(rows) + 1):
            try:
                reference = contacts_df.iloc[row_index - 1]['Reference']
                if reference not in existing_data['Reference'].values:
                    combined_df = process_row(driver, row_index, contacts_df, wait)
                    all_combined_dfs.append(combined_df)
                    existing_data = pd.concat([existing_data, combined_df], ignore_index=True)
                    existing_data.to_excel('final_combined_data.xlsx', index=False)
                    print(f"Returning to Contacts page after processing row {row_index} on page {page_number}...")
                else:
                    print(f"Skipping row {row_index} on page {page_number} as it is already processed.")
            except Exception as e:
                print(f"Error processing row {row_index} on page {page_number}: {e}")
                continue
        
        try:
            print("Clicking on the next page button...")
            next_button = wait.until(EC.element_to_be_clickable((By.XPATH, '//*[contains(@id, "DataTables_Table") and contains(@id, "_paginate")]/ul/li[@class="next"]/a')))
            driver.execute_script("arguments[0].click();", next_button)
            print("Waiting for the next page to load...")
            page_number += 1
            time.sleep(5)  # Adjust this sleep time as needed
        except Exception as e:
            print("No more pages or an error occurred:", e)
            break
finally:
    print("Saving inner pages DataFrame to an Excel file...")
    if inner_pages:
        inner_pages_df = pd.concat(inner_pages, axis=0, ignore_index=True)
        inner_pages_df.to_excel('inner_pages_data.xlsx', index=False)

    print("Saving folio pages DataFrame to an Excel file...")
    if folio_pages:
        folio_pages_df = pd.concat(folio_pages, axis=0, ignore_index=True)
        folio_pages_df.to_excel('folio_pages_data.xlsx', index=False)

    print("Saving transactions DataFrame to an Excel file...")
    if transaction_pages:
        transaction_pages_df = pd.concat(transaction_pages, axis=0, ignore_index=True)
        transaction_pages_df.to_excel('transactions_data.xlsx', index=False)

    print("Concatenating all combined DataFrames...")
    if all_combined_dfs:
        final_combined_df = pd.concat(all_combined_dfs, axis=0, ignore_index=True)
        print("Saving final combined DataFrame to a new Excel file...")
        final_combined_df.to_excel('final_combined_data.xlsx', index=False)

    print("Closing the driver...")
    driver.quit()
```
